# 2. Ventas por categoría

```python
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
# calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría
#  y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

########################## conteo tienda 1 #####################################

# ejemplo en el archivo  "tienda.csv" hay la columna: Producto, categoria_producto 
tienda = pd.read_csv(url, usecols=['Producto', 'Categoría del Producto'])
# Contar cuántos productos se vendieron por categoría
conteo_tienda_1 = tienda.groupby('Categoría del Producto').count().reset_index()
# Renombrar la columna para que sea más clara
conteo_tienda_1 = conteo_tienda_1.rename(columns={'Producto': 'cantidad_vendida'})

########################## conteo tienda 2 #####################################

# ejemplo en el archivo  "tienda.csv" hay la columna: Producto, categoria_producto 
tienda_2 = pd.read_csv(url2, usecols=['Producto', 'Categoría del Producto'])
# Contar cuántos productos se vendieron por categoría
conteo_tienda_2 = tienda_2.groupby('Categoría del Producto').count().reset_index()
# Renombrar la columna para que sea más clara
conteo_tienda_2 = conteo_tienda_2.rename(columns={'Producto': 'cantidad_vendida'})

########################## conteo tienda 3 #####################################

# ejemplo en el archivo  "tienda.csv" hay la columna: Producto, categoria_producto 
tienda_3 = pd.read_csv(url3, usecols=['Producto', 'Categoría del Producto'])
# Contar cuántos productos se vendieron por categoría
conteo_tienda_3 = tienda_3.groupby('Categoría del Producto').count().reset_index()
# Renombrar la columna para que sea más clara
conteo_tienda_3 = conteo_tienda_3.rename(columns={'Producto': 'cantidad_vendida'})

########################## conteo tienda 4 #####################################

# ejemplo en el archivo  "tienda.csv" hay la columna: Producto, categoria_producto 
tienda_4 = pd.read_csv(url4, usecols=['Producto', 'Categoría del Producto'])
# Contar cuántos productos se vendieron por categoría
conteo_tienda_4 = tienda_4.groupby('Categoría del Producto').count().reset_index()
# Renombrar la columna para que sea más clara
conteo_tienda_4 = conteo_tienda_4.rename(columns={'Producto': 'cantidad_vendida'})

print("*********** Ventas por categoria de la tienda 1 ************")
print(conteo_tienda_1)
print("*********** Ventas por categoria de la tienda 2 ************")
print(conteo_tienda_2)
print("*********** Ventas por categoria de la tienda 3 ************")
print(conteo_tienda_3)
print("*********** Ventas por categoria de la tienda 4 ************")
print(conteo_tienda_4)
```

# 3. Calificación promedio de la tienda


```python
# Debes calcular las calificaciones promedio de los clientes para cada tienda. 
# El objetivo es conocer la satisfacción del cliente con los productos vendidos.

tienda = pd.read_csv(url, usecols=['Calificación'])
tienda_2 = pd.read_csv(url2, usecols=['Calificación'])
tienda_3 = pd.read_csv(url3, usecols=['Calificación'])
tienda_4 = pd.read_csv(url4, usecols=['Calificación'])
# Calcular el promedio de la columna 'calificacion'
promedio = tienda['Calificación'].mean()
promedio_2 = tienda_2['Calificación'].mean()
promedio_3 = tienda_3['Calificación'].mean()
promedio_4 = tienda_4['Calificación'].mean()

print(f"Promedio de calificaciones de los clientes para la tienda 1: {promedio}")
print(f"Promedio de calificaciones de los clientes para la tienda 2: {promedio_2}")
print(f"Promedio de calificaciones de los clientes para la tienda 3: {promedio_3}")
print(f"Promedio de calificaciones de los clientes para la tienda 4: {promedio_4}")
```

# 4. Productos más y menos vendidos

```python
# debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. 
# Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda
# ejemplo en el archivo  "tienda.csv" hay la columna: Producto 

print("*********** Producto mas y menos vendido de la tienda 1 ************")
tienda = pd.read_csv(url, usecols=['Producto'])
# Contar cuántas veces se vendió cada producto
conteo = tienda['Producto'].value_counts()
print(conteo)
# Obtener el producto más vendido
mas_vendido = conteo.idxmax()
cant_mas_vendido = conteo.max()

# Obtener el producto menos vendido
menos_vendido = conteo.idxmin()
cant_menos_vendido = conteo.min()

print(f"🥇 Producto más vendido: {mas_vendido} ({cant_mas_vendido} ventas)")
print(f"🔻 Producto menos vendido: {menos_vendido} ({cant_menos_vendido} venta/s)")
```
